```python
# Imports
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # plotting
import plotly.express as px # plotting

#Reading file through read_csv method
df = pd.read_csv('/kaggle/input/sales-forecasting/train.csv')

# Change date to a more readable format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Filter, leaving only Segment, Date by Month for the first year, and Total Sales, then rename accordingly
month_wise_sale = df.groupby(['Segment',df['Order Date'].dt.month])['Sales'].sum().reset_index()
month_wise_sale = month_wise_sale.rename(columns={'Order Date': 'Month', 'Sales':'Total Sales'})

# Manual separating of segments
a = month_wise_sale['Month'][0:12]
b = month_wise_sale['Total Sales'][0:12]

c = month_wise_sale['Month'][12:24]
d = month_wise_sale['Total Sales'][12:24]

e = month_wise_sale['Month'][24:36]
f = month_wise_sale['Total Sales'][24:36]

# Plotting 
plt.plot(a, b, color='red', linestyle='-',label='Consumer')
plt.plot(c, d, color='blue', linestyle='-',label='Corporate')
plt.plot(e, f, color='green', linestyle='-',label='Home Office')

# Labeling
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')
plt.tight_layout()
plt.legend()
```

```python
##Next one


# Filter, leaving only Segment, Date by Year, and Total Sales, then rename accordingly
year_wise_sale = df.groupby(['Segment',df['Order Date'].dt.year])['Sales'].sum().reset_index()
year_wise_sale = year_wise_sale.rename(columns={'Order Date': 'Year', 'Sales':'Total Sales'})

# Manual separating of segments
a = year_wise_sale['Year'][0:4]
b = year_wise_sale['Total Sales'][0:4]

c = year_wise_sale['Year'][4:8]
d = year_wise_sale['Total Sales'][4:8]

e = year_wise_sale['Year'][8:12]
f = year_wise_sale['Total Sales'][8:12]

# Plotting
plt.plot(a, b, color='red', linestyle='-',label='Consumer')
plt.plot(c, d, color='blue', linestyle='-',label='Corporate')
plt.plot(e, f, color='green', linestyle='-',label='Home Office')

# Labeling
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')
plt.tight_layout()
plt.legend()
```
