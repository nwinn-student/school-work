```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.express as px

df = pd.read_csv("/kaggle/input/indian-census-data-with-geospatial-indexing/district wise population and centroids.csv")
df.head()
```

```python
df.info() # no nulls
```

```python
df.describe()
```

```python
change = df['Population in 2011'] - df['Population in 2001']
zeroChange = []
for i in range(len(change)):
    if change[i] > 0:
        zeroChange.append(change[i])
    else:
        zeroChange.append(0)
```

```python
df.insert(5, 'Change in Population',change, True)
df.insert(6, 'Zero"d Change in Population', zeroChange, True)
df
```

```python
fig = px.scatter_mapbox(
    df, lat = 'Latitude', 
    lon = 'Longitude', 
    zoom = 3, 
    hover_name = 'District',
    hover_data = 'Change in Population',
    size = 'Zero"d Change in Population',
    mapbox_style = 'carto-positron'
)
fig.show()
```
