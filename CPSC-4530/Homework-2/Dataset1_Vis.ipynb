```python
import pandas as pd
import plotly.express as px
import math

df = pd.read_csv('/kaggle/input/crop-production/Production_Crops_E_Oceania.csv')

df
```

```python
df['Area'].nunique() # 18
df['Item'].nunique()
```

```python
production = []
summed = 0
p = df.groupby(['Item','Element'])[[column for column in df.columns if column.startswith('Y') and not column.endswith('F')]].sum()
p
```

```python
# initializes list
production = []
for x in p:
    for v in range(math.ceil(len(p[x]) / 3)+1):
            production.append(0)
    break
    
# ensures only production values cross year are summed
i = 0
for x in p:
    for v in range(math.floor((len(p[x]) / 3))+1):
        i = v
        production[v+1] += p[x][v*3]
    # print(p[x][(i-98)*3])
    # print(production[i-98])
    production[i-98] += p[x][(i-98)*3]
production
```

```python
# They must be equal
print(len(production))
print(len(df['Item'].unique()))

summer = 0
for x in production:
    summer += x
summer
```

```python
uniqueItems = pd.DataFrame(df['Item'].unique())
uniqueItems = uniqueItems.sort_values(by=0)
print(len(uniqueItems.to_numpy()))

items = []
for x in range(len(uniqueItems.to_numpy())):
    items.append(uniqueItems.to_numpy()[x][0])
# items
```

```python
df['Item'].unique()
```

```python
fig = px.histogram(
    data_frame=df, 
    x=items,
    y=production,
    range_x=(0,10),
    range_y=(50000000,2000000000),
    log_y=True,
    title='Top 10 Oceania Crops from 1961-2019',
).update_xaxes(categoryorder = "total descending")

fig.update_layout(
    xaxis_title="Crops",
    yaxis_title="Production Amount (tonnes)",
)

fig.show()
```
